```python
import matplotlib.pyplot as plt

# Calculate the absolute difference
df['diff'] = (df['mean123'] - df['mean456']).abs()

# Sort and select top 100
df_sorted = df.sort_values('diff', ascending=False)
top100 = df_sorted.head(100)

# Use the index (IlmnID) for labels
plt.figure(figsize=(20, 6))
plt.bar(top100.index.astype(str), top100['diff'])
plt.xlabel('IlmnID')
plt.ylabel('Absolute Difference')
plt.title('Top 100 Biggest Differences Between mean123 and mean456')
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

cols_123 = ['GSM3193366_semen1', 'GSM3193367_semen2', 'GSM3193368_semen3']
cols_456 = ['GSM3193369_semen4', 'GSM3193370_semen5', 'GSM3193371_semen6']

# Plot for top100_gt
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
for col in cols_123:
    top100_gt[col].plot(kind='kde', color='blue', alpha=0.6, label='123' if col == cols_123[0] else "")
for col in cols_456:
    top100_gt[col].plot(kind='kde', color='red', alpha=0.6, label='456' if col == cols_456[0] else "")
plt.title('Density Plot: mean123 > mean456')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Plot for top100_lt
plt.subplot(1, 2, 2)
for col in cols_123:
    top100_lt[col].plot(kind='kde', color='blue', alpha=0.6, label='123' if col == cols_123[0] else "")
for col in cols_456:
    top100_lt[col].plot(kind='kde', color='red', alpha=0.6, label='456' if col == cols_456[0] else "")
plt.title('Density Plot: mean123 < mean456')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate the signed difference
top100['diff_signed'] = top100['mean123'] - top100['mean456']

# Set colors: red for negative, blue for positive
colors = np.where(top100['diff_signed'] < 0, 'red', 'blue')

plt.figure(figsize=(20, 6))
plt.bar(top100.index.astype(str), top100['diff_signed'], color=colors)
plt.xlabel('IlmnID')
plt.ylabel('mean123 - mean456')
plt.title('Top 100 Differences Between mean123 and mean456 (Red = Negative)')
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```
